```python
thetasb, lossesb = lr.lr2(X_data_b, y_data_b)
```

```python
fig, axes = plt.subplots(2, 3, figsize = (16,8))
axes = axes.ravel()

for k , theta in enumerate(thetasb[-6:]):
    ax = axes[k]
    xs = np.arange(0, 1.1, 0.1)
    ys = -(theta[0] + theta[1] * xs) / theta[2]
    ax.plot(xs, ys, lw = 0.5)
    data_b.query('label == -1.').plot.scatter(x = 'x1', y = 'x2', ax = ax, color = 'blue')
    data_b.query('label == 1.').plot.scatter(x = 'x1', y = 'x2', ax = ax, color = 'red')
```

由上面可知，data_b的分类总是趋于将误差近似到0, 也就是将theta趋于无穷大，但是这个在implementation的时候是无法做到的。  
如果说用SVM，那么这种问题可以得到缓解。  
SVM能够达到缓解的根本原因是SVM自带正则，它是在最小边界的情况下最小化权重。所以不会出现想要把theta优化到无穷大的情况。

i. 如果用另外的常数learning_rate也不会环节，因为LR是为了最大化最大似然，也就是让theta区域无穷大。如果说learning_rate很大的时候，可能达到数值的无穷大(例如1e15), 但是还是无法让data_b的数据拟合尽快收敛。

ii.Decreasing the learning rate over time (e.g. scaling the initial learning rate by $1/t^2$, where $t$ is the number of gradient descent iterations thus far).  
这种处理方式在这里无效，这不是因为后面的优化过程当中learning_rate过大导致优化震荡的问题。  

iii.Adding a regularization term $\left|\left|\theta\right|\right|^2_2$ to the loss function. 
有用，因为正则项在这里对theta进行惩罚，会让误差不逼近零，对theta的值进行限制。所以模型不会想要最大化theta。

iv. Linear scaling of the input features.  
无效，因为这个与输入的参数没有关系。  

v. Adding zero-mean Gaussian noise to the training data or labels.  
不一定有用，如果数据还是线性可分的，模型还是会区域优化theta到无穷大。
